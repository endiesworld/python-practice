#### Adding new columns
A more efficient way to add new columns to an existing Dataframe is to use the apply() method.
**apply()** method also accepts a call back function, which is called on every iteration. The return value of the function is assigned a the value for the new column.

```python
def price_2024(old_price):
    return old_price * 1.25
```

```python
first_ten_rows['2024_price'] = first_ten_rows['price'].apply(price_2024)
first_ten_rows
```
