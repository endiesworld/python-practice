## Scatter Plot
Also known as a Scatter Graph, Point Graph, X-Y Plot, Scatter Chart or Scattergram.

A Scatterplot places points on a Cartesian Coordinates system to display all the values between two variables. By having an axis for each variable, you can detect if a relationship or correlation between the two exist
The kind of correlation can be interpreted through the patterns revealed on a Scatterplot. These are: positive (values increase together), negative (one value decreases as the other increases) or null (no correlation). The shape of the correlation can be described as: linear, exponential and U-shaped. The strength of the correlation can be determined by how closely packed the points are to each other on the graph. Points that end up far outside the general cluster of points are known as outliers.s.

```python
# Change the line plot to a scatter plot.
# A correlation will become clear when you display the GDP per capita on a logarithmic scale. Add the line plt.xscale('log')
plt.scatter(gdp_cap, life_exp)
plt.show()
```

### logarithmic scale
Why put the x-axis on a logarithmic scale?

#### Assistant
Putting the x-axis on a logarithmic scale can be beneficial in several scenarios:

1. **Handling Wide Ranges**: When the data spans several orders of magnitude, a logarithmic scale can help visualize the data more clearly. For example, if you have values ranging from 1 to 1,000,000, a linear scale would compress most of the data into a small area, making it hard to interpret.

2. **Exponential Growth**: If the data represents exponential growth (e.g., population growth, compound interest), a logarithmic scale can linearize the data, making trends easier to identify and analyze.

3. **Proportional Relationships**: Logarithmic scales can help in visualizing proportional relationships. For instance, if you are comparing rates of change, a log scale can show relative changes more effectively than a linear scale.

4. **Reducing Skewness**: Logarithmic transformation can reduce skewness in data distributions, making it easier to apply statistical methods that assume normality.

5. **Improved Clarity**: In some cases, a log scale can help highlight smaller values that might be lost in a linear scale, providing a clearer picture of the data distribution.

In summary, using a logarithmic scale on the x-axis can enhance the interpretability of data, especially when dealing with exponential relationships or wide-ranging values.

#### what is skewness in data ?

Skewness is a statistical measure that describes the asymmetry of a probability distribution. It indicates the extent and direction of deviation from the normal distribution (which is symmetric). Here are the key points regarding skewness:

1. **Types of Skewness**:
   - **Positive Skewness (Right Skew)**: The tail on the right side of the distribution is longer or fatter than the left side. In this case, the mean is typically greater than the median.
   - **Negative Skewness (Left Skew)**: The tail on the left side is longer or fatter than the right side. Here, the mean is usually less than the median.
   - **Zero Skewness**: This indicates a perfectly symmetrical distribution, such as a normal distribution, where the mean and median are equal.

2. **Interpretation**:
   - A positive skewness indicates that most of the data points are concentrated on the left side of the distribution, with a few larger values pulling the mean to the right.
   - A negative skewness suggests that most data points are concentrated on the right side, with a few smaller values pulling the mean to the left.

3. **Calculation**: Skewness can be calculated using various formulas, but one common method is Pearson's moment coefficient of skewness. The formula is:
   \[
   \text{Skewness} = \frac{n}{(n-1)(n-2)} \sum \left( \frac{x_i - \bar{x}}{s} \right)^3
   \]
   where \( n \) is the number of observations, \( x_i \) are the data points, \( \bar{x} \) is the mean, and \( s \) is the standard deviation.

4. **Importance**: Understanding skewness is crucial in statistics and data analysis because it affects the choice of statistical tests and the interpretation of data. Many statistical methods assume normality, and skewed data may require transformation or different analytical approaches.

In summary, skewness provides insight into the shape of the data distribution, helping analysts understand the underlying patterns and make informed decisions based on the data.
