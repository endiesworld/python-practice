```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
house_data_pd = pd.read_csv('housing_data.csv')
print(house_data_pd.shape)
house_data_pd.head()
```

```python
print(house_data_pd.columns)
```

```python
# To review how the SalesPrice colunm is distributed
house_data_pd['SalePrice'].describe()
```

```python
# plots the distribution of the SalesPrice column
sns.displot(house_data_pd['SalePrice'])
plt.show()
plt.clf()
```

```python
plt.hist(house_data_pd['SalePrice'])
plt.show()
```

```python
# perform log transformation on sale price data
log_data = np.log(house_data_pd['SalePrice'])
print(log_data.skew())
sns.displot(log_data)
plt.show()
plt.clf()
```

```python
plt.hist(log_data)
plt.show()
```

```python
# visualize relationship between 1st floor sq feet and sale price
data = pd.concat([house_data_pd['SalePrice'], house_data_pd['1stFlrSF']], axis=1)
data.plot.scatter(x='1stFlrSF', y='SalePrice', ylim=(0,800000))
plt.show()
plt.clf()
```

```python
# visualize relationship between garage area and sale price
data = pd.concat([house_data_pd['SalePrice'], house_data_pd['GarageArea']], axis=1)
data.plot.scatter(x='GarageArea', y='SalePrice', ylim=(0,800000))
plt.show()
plt.clf()
```

```python
# visualize relationship between year built and sale price
data = pd.concat([house_data_pd['SalePrice'], house_data_pd['YearBuilt']], axis=1)
f, ax = plt.subplots(figsize=(16, 8))
fig = sns.boxplot(x='YearBuilt', y="SalePrice", data=data)
fig.axis(ymin=0, ymax=800000)
plt.xticks(rotation=90)
plt.show()
plt.clf()
```
